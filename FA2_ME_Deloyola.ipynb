```r
library(tidyr)
library(dplyr)
library(tidyverse)
library(ggplot2)
```

(a) Construct a scatter plot for the Height vs Arm.

```r
body <- read.csv("/content/body-2.csv")
x <- body$Height
y <- body$Arm

head(body)
plot(x,y,
     col = "orange",
     pch = 20,
     main = "Height vs Arm",
     xlab = "Height",
     ylab = "Arm")

plot(body,
     col = "orange",
     pch = 20,
     xlab = "Arm",
     ylab = "Height")
```

(b) Formulate the least square linear model for Height vs Arm and report the summary table.

```r
body_lm <- lm(y~x, data=body)

plot(x=x, y=y,
     xlab = "Height",
     ylab = "Arm",
     pch=18,
     col="darkgreen") #scatter plot for x and y

abline(body_lm,
      col="pink",
      lwd=5) #abline for linear model

summary(body)
```

(c) Evaluate the model by looking at the regression standard error, R^2, and b1.

```r
body.regression <- lm(y~x, data = body)
summary(body.regression)
```
